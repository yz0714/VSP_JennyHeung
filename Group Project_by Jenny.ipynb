## Comparison of Key Economic Indicators

### 1. Gross domestic product

Gross domestic product (GDP) is the standard measure of the value added created through the production of goods and services in a country during a certain period.

```python
import pandas as pd
import matplotlib.pyplot as plt
aus_gdp_data = pd.read_csv('GDP of Australia.csv')
nz_gdp_data = pd.read_csv('GDP of New Zealand.csv')

aus_gdp = aus_gdp_data[' Per Capita (US $)']
nz_gdp = nz_gdp_data[' Per Capita (US $)']

years = aus_gdp_data['Year']

combined_gdp_data = pd.DataFrame({'Australia': aus_gdp, 'New Zealand': nz_gdp})
```

```python
combined_gdp_data.plot(marker='o')
plt.title('GDP per capita of Australia and New Zealand')
plt.xlabel('years')
plt.ylabel('GDP per Capita (US $)')
plt.legend()
plt.xticks(range(len(years)), years, rotation=45)
plt.show()
```

### 2. Gross national income

Gross national income (GNI) is defined as gross domestic product, plus net receipts from abroad of compensation of employees, property income and net taxes less subsidies on production.

```python
aus_gni_data = pd.read_csv('GNI of Australia.csv')
nz_gni_data = pd.read_csv('GNI of New Zealand.csv')

aus_gni = aus_gni_data[' GNI Per Capita (US $)']
nz_gni = nz_gni_data[' GNI Per Capita (US $)']

years = aus_gni_data['Year']

combined_gni_data = pd.DataFrame({'Australia': aus_gni, 'New Zealand': nz_gni})
```

```python
combined_gni_data.plot(marker='o')
plt.title('GNI per capita of Australia and New Zealand')
plt.xlabel('years')
plt.ylabel('GNI Per Capita (US $)')
plt.legend()
plt.xticks(range(len(years)), years, rotation=45)
plt.show()
```

### 3. Inflation rates

Inflation is a rise in prices, which can be translated as the decline of purchasing power over time. The rate at which purchasing power drops can be reflected in the average price increase of a basket of selected goods and services over some period of time.

```python
aus_inflation_data = pd.read_csv('Inflation rates for Australia.csv', dtype={'Inflation rates(%)': float})
nz_inflation_data = pd.read_csv('Inflation rates for New Zealand.csv', dtype={'Inflation rates(%)': float})

aus_inflation = aus_inflation_data['Inflation rates(%)']
nz_inflation = nz_inflation_data['Inflation rates(%)']

years = aus_inflation_data['Year']

combined_inflation_data = pd.DataFrame({'Australia': aus_inflation, 'New Zealand': nz_inflation})
```

```python
combined_inflation_data.plot(marker='o')
plt.title('Inflation rates of Australia and New Zealand')
plt.xlabel('Years')
plt.ylabel('Inflation rates(%)')
plt.legend()
plt.xticks(range(len(years)), years, rotation=45)
plt.show()
```

```python
aus_unemployment_data = pd.read_csv('Unemployment Rate of Australia.csv', dtype={'Unemployment Rate (%)': float})
nz_unemployment_data = pd.read_csv('Unemployment Rate of New Zealand.csv', dtype={'Unemployment Rate (%)': float})

aus_unemployment = aus_unemployment_data['Unemployment Rate (%)']
nz_unemployment = nz_unemployment_data['Unemployment Rate (%)']

years = aus_unemployment_data['Year']

combined_unemployment_data = pd.DataFrame({'Australia': aus_unemployment, 'New Zealand': nz_unemployment})
```

```python
combined_unemployment_data.plot(marker='o')
plt.title('Unemployment rates of Australia and New Zealand')
plt.xlabel('Years')
plt.ylabel('Unemployment Rate (%)')
plt.legend()
plt.xticks(range(len(years)), years, rotation=45)
plt.show()
```

### 5. House Prices Growth

Housing prices include housing rent prices indices, real and nominal house prices indices, and ratios of price to rent and price to income.

```python
aus_hseprice_data = pd.read_csv('House Prices Growth of Australia.csv', dtype={'House Prices Growth (%)': float})
nz_hseprice_data = pd.read_csv('House Prices Growth of New Zealand.csv', dtype={'House Prices Growth (%)': float})

aus_hseprice = aus_hseprice_data['House Prices Growth (%)']
nz_hseprice = nz_hseprice_data['House Prices Growth (%)']

years = aus_hseprice_data['Year']

combined_hseprice_data = pd.DataFrame({'Australia': aus_hseprice, 'New Zealand': nz_hseprice})
```

### 6. Labour Force Participation Rate

The labour force participation rates is calculated as the labour force divided by the total working-age population. The working age population refers to people aged 15 to 64.

```python
aus_lfp_data = pd.read_csv('Labour Force Participation Rate of Australia.csv', dtype={'Participation Rate (%)': float})
nz_lfp_data = pd.read_csv('Labour Force Participation Rate of New Zealand.csv', dtype={'Participation Rate (%)': float})

aus_lfp = aus_lfp_data['Participation Rate (%)']
nz_lfp = nz_lfp_data['Participation Rate (%)']

years = aus_lfp_data['Year']

combined_lfp_data = pd.DataFrame({'Australia': aus_lfp, 'New Zealand': nz_lfp})
```

### 7. Foreign Direct Investment

Foreign direct investment (FDI) is a category of cross-border investment in which an investor resident in one economy establishes a lasting interest in and a significant degree of influence over an enterprise resident in another economy.

```python
aus_fdi_data = pd.read_csv('Foreign Direct Investment of Australia.csv')
nz_fdi_data = pd.read_csv('Foreign Direct Investment of New Zealand.csv')

aus_fdi = aus_fdi_data['Foreign Direct Investment in Dec (USD in million)']
nz_fdi = nz_fdi_data['Foreign Direct Investment in Dec (USD in million)']

years = aus_fdi_data['Year']

combined_fdi_data = pd.DataFrame({'Australia': aus_fdi, 'New Zealand': nz_fdi})
```

```python
combined_lfp_data.plot(marker='o')
plt.title('Foreign Direct Investment of Australia and New Zealand')
plt.xlabel('Years')
plt.ylabel('Foreign Direct Investment in Dec (USD in million)')
plt.legend()
plt.xticks(range(len(years)), years, rotation=45)
plt.show()
```

### 8. Currency Exchange Rate (to USD)

Exchange rates are defined as the price of one country's' currency in relation to another country's currency.

```python
aus_exchange_data = pd.read_csv('USD-ANU.csv')
nz_exchange_data = pd.read_csv('USD-NZD.csv')

aus_exchange = aus_exchange_data['USD/AUD']
nz_exchange = nz_exchange_data['USD/NZD']

years = aus_exchange_data['Year']

combined_exchange_data = pd.DataFrame({'Australia': aus_exchange, 'New Zealand': nz_exchange})
```

```python
plt.plot(years, aus_exchange, marker='o', label='Australia')
plt.plot(years, nz_exchange, marker='o', label='New Zealand')

plt.title('Exchange Rates of Australia and New Zealand')
plt.xlabel('Years')
plt.ylabel('Exchange Rate')

plt.legend()
plt.xticks(rotation=45)
plt.show()
```

## Immigration situation of Australia and New Zealand

```python
aus_migration_data = pd.read_csv('Net Migration Rate of Australia.csv')
nz_migration_data = pd.read_csv('Net Migration Rate of New Zealand.csv')

aus_migration = aus_migration_data[' Per 1000 Population']
nz_migration = nz_migration_data[' Per 1000 Population']

years = aus_migration_data['Year']

combined_migration_data = pd.DataFrame({'Australia': aus_migration, 'New Zealand': nz_migration})
```

```python
combined_migration_data.plot(kind = 'bar', width = 0.8)
plt.title('Net Migration Rate of Australia and New Zealand')
plt.xlabel('Years')
plt.ylabel('Net Migration Rate Per 1000 Population')
plt.legend()
plt.xticks(range(len(years)), years, rotation=45)
plt.show()
```

## Relationship between immigration rates and economic indicators

```python
import seaborn as sns
import statsmodels.api as sm

aus_relationship_data = pd.read_csv('indicator relationship of Australia.csv')

plt.figure(figsize=(10, 6))
sns.scatterplot(x=aus_relationship_data['GDP(Billions of US $)'], y=aus_relationship_data['Net migration Per 1000 Population'])
plt.title('Scatter Plot: GDP vs Net Migration in Australia')
plt.xlabel('GDP (Billions of US $)')
plt.ylabel('Net Migration Per 1000 Population')
plt.show()
```

In most cases, the higher the GDP, the higher the immigration rate.

```python
aus_relationship_data = pd.read_csv('indicator relationship of New Zealand.csv')

plt.figure(figsize=(10, 6))
sns.scatterplot(x=aus_relationship_data['GDP(Billions of US $)'], y=aus_relationship_data['Net migration Per 1000 Population'])
plt.title('Scatter Plot: GDP vs Net Migration in New Zealand')
plt.xlabel('GDP (Billions of US $)')
plt.ylabel('Net Migration Per 1000 Population')
plt.show()
```

In most cases, the higher the GDP, the lower the immigration rate.

## Conclusion

Economic indicators in Australia are more closely related to immigration rates. New Zealand's economic indicators have little impact on immigration rates.
